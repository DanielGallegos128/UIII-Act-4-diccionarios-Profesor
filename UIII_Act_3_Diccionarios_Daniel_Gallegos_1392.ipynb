### Diccionarios en Python

```python
# Inicialización del diccionario vacío
diccionario_profesor = {}

# --- PRE-CARGA DE 4 DATOS INICIALES ---
diccionario_profesor["Carlos Mendoza"] = {"Materia": "Matemáticas", "Grado Académico": "Maestría", "Institución": "UANL", "Alumnos a cargo": 120, "Nivel Educativo": "Universitario", "Años de docencia": 10}
diccionario_profesor["Ana Rodríguez"] = {"Materia": "Historia", "Grado Académico": "Licenciatura", "Institución": "Colegio Hidalgo", "Alumnos a cargo": 45, "Nivel Educativo": "Secundaria", "Años de docencia": 5}
diccionario_profesor["Luis Fuentes"] = {"Materia": "Física", "Grado Académico": "Doctorado", "Institución": "UNAM", "Alumnos a cargo": 80, "Nivel Educativo": "Universitario", "Años de docencia": 15}
diccionario_profesor["Elena Gómez"] = {"Materia": "Química", "Grado Académico": "Maestría", "Institución": "Prepa Tec", "Alumnos a cargo": 60, "Nivel Educativo": "Preparatoria", "Años de docencia": 8}

# --- FUNCION AUXILIAR PARA VALIDAR NÚMEROS ENTEROS ---
def solicitar_entero(mensaje):
    """Sigue pidiendo el dato interactivamente hasta que sea un número entero válido."""
    while True:
        valor = input(mensaje)
        if valor.isdigit():
            return int(valor)
        print("❌ Error: Por favor, introduce un número entero válido.")

# --- FUNCIONES SOLICITADAS ---
def agregar_profesor():
    print("\n--- ➕ AGREGAR NUEVO PROFESOR ---")
    nombre = input("Nombre del profesor: ").strip()

    if not nombre:
        print("❌ El nombre no puede estar vacío.")
        return
    if nombre in diccionario_profesor:
        print("⚠ ¡Este profesor ya existe! Usa la opción de actualizar.")
        return

    materia = input("Materia impartida: ")
    grado = input("Grado académico: ")
    institucion = input("Institución: ")
    alumnos = solicitar_entero("Cantidad de alumnos a cargo: ")
    nivel = input("Nivel educativo: ")
    anios_docencia = solicitar_entero("Años de docencia: ")

    diccionario_profesor[nombre] = {
        "Materia": materia,
        "Grado Académico": grado,
        "Institución": institucion,
        "Alumnos a cargo": alumnos,
        "Nivel Educativo": nivel,
        "Años de docencia": anios_docencia
    }
    print(f"🎉 ¡Profesor '{nombre}' agregado con éxito!")


def listar_profesor():
    print("\n--- 📋 LISTA DE PROFESORES ---")
    if not diccionario_profesor:
        print("No hay profesores registrados en el sistema.")
        return

    # Columnas con anchos fijos para evitar deformaciones
    ancho_nom, ancho_mat, ancho_grad, ancho_inst, ancho_alum, ancho_niv, ancho_anio = 20, 15, 15, 18, 10, 15, 10

    formato_fila = f"| {{:<{ancho_nom}}} | {{:<{ancho_mat}}} | {{:<{ancho_grad}}} | {{:<{ancho_inst}}} | {{:<{ancho_alum}}} | {{:<{ancho_niv}}} | {{:<{ancho_anio}}} |"
    linea_divisoria = "-" * (ancho_nom + ancho_mat + ancho_grad + ancho_inst + ancho_alum + ancho_niv + ancho_anio + 22)

    encabezados = ["Nombre", "Materia", "Grado Acad.", "Institución", "Alumnos", "Nivel Ed.", "Años Doc."]

    print(linea_divisoria)
    print(formato_fila.format(*encabezados))
    print(linea_divisoria)

    # REQUISITO: Uso de for key, value (nombre, datos)
    for nombre, datos in diccionario_profesor.items():
        print(formato_fila.format(
            nombre[:ancho_nom],
            datos["Materia"][:ancho_mat],
            datos["Grado Académico"][:ancho_grad],
            datos["Institución"][:ancho_inst],
            str(datos["Alumnos a cargo"])[:ancho_alum],
            datos["Nivel Educativo"][:ancho_niv],
            str(datos["Años de docencia"])[:ancho_anio]
        ))
    print(linea_divisoria)


def buscar_profesor():
    print("\n--- 🔍 BUSCAR PROFESOR ---")
    nombre = input("Ingrese el nombre del profesor a buscar: ").strip()

    if nombre in diccionario_profesor:
        print(f"\n✅ Registro encontrado para: {nombre}")
        # REQUISITO: Uso de for key, value para mostrar las características
        for caracteristica, detalle in diccionario_profesor[nombre].items():
            print(f"   • {caracteristica}: {detalle}")
    else:
        print("❌ Profesor no encontrado en los registros.")


def borrar_profesor():
    print("\n--- 🗑 BORRAR PROFESOR ---")
    nombre = input("Ingrese el nombre del profesor a eliminar: ").strip()

    if nombre in diccionario_profesor:
        del diccionario_profesor[nombre]
        print(f"🗑 El profesor '{nombre}' ha sido eliminado correctamente.")
    else:
        print("❌ Profesor no encontrado.")


def actualizar_profesor():
    print("\n--- 🔄 ACTUALIZAR PROFESOR ---")
    nombre = input("Ingrese el nombre del profesor que desea actualizar: ").strip()

    if nombre in diccionario_profesor:
        print("💡 (Presiona [ENTER] sin escribir nada para conservar el dato actual)")

        # Guardamos una copia del registro actual para facilitar el código
        actual = diccionario_profesor[nombre]

        materia = input(f"Materia [{actual['Materia']}]: ").strip() or actual['Materia']
        grado = input(f"Grado académico [{actual['Grado Académico']}]: ").strip() or actual['Grado Académico']
        institucion = input(f"Institución [{actual['Institución']}]: ").strip() or actual['Institución']

        # Validación interactiva para números en la actualización
        while True:
            alumnos_in = input(f"Alumnos a cargo [{actual['Alumnos a cargo']}]: ").strip()
            if not alumnos_in:
                alumnos = actual['Alumnos a cargo']
                break
            elif alumnos_in.isdigit():
                alumnos = int(alumnos_in)
                break
            print("❌ Debe ser un número entero.")

        nivel = input(f"Nivel educativo [{actual['Nivel Educativo']}]: ").strip() or actual['Nivel Educativo']

        while True:
            anios_in = input(f"Años de docencia [{actual['Años de docencia']}]: ").strip()
            if not anios_in:
                anios_docencia = actual['Años de docencia']
                break
            elif anios_in.isdigit():
                anios_docencia = int(anios_in)
                break
            print("❌ Debe ser un número entero.")

        # Guardar cambios
        diccionario_profesor[nombre] = {
            "Materia": materia,
            "Grado Académico": grado,
            "Institución": institucion,
            "Alumnos a cargo": alumnos,
            "Nivel Educativo": nivel,
            "Años de docencia": anios_docencia
        }
        print(f"✅ ¡Datos de '{nombre}' actualizados con éxito!")
    else:
        print("❌ Profesor no encontrado.")


# --- MENÚ INTERACTIVO PRINCIPAL (While True + If/Elif) ---
while True:
    print("\n=================== MENÚ PRINCIPAL ===================")
    print(" 1. Agregar profesor en un diccionario")
    print(" 2. Listar diccionarios de profesores (Tabla)")
    print(" 3. Buscar profesor en diccionario")
    print(" 4. Borrar profesor en diccionario")
    print(" 5. Actualizar profesor en diccionario")
    print(" 6. Salir")
    print("======================================================")

    opcion = input("Seleccione una opción (1-6): ").strip()

    if opcion == "1":
        agregar_profesor()
    elif opcion == "2":
        listar_profesor()
    elif opcion == "3":
        buscar_profesor()
    elif opcion == "4":
        borrar_profesor()
    elif opcion == "5":
        actualizar_profesor()
    elif opcion == "6":
        print("\n👋 Saliendo del sistema interactivo. ¡Buen día!")
        break
    else:
        print("❌ Opción inválida. Por favor, digite un número del 1 al 6.")
```
